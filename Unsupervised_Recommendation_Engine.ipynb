```python
skills_lst = [
    'tableau', 'data warehousing',
    'machine learning', 'etl', 'power bi', 
    'scala', 'excel', 'hadoop', 'oracle', 
    'r', 'data pipeline', 'spark', 'sql', 
    'data governance', 'data engineering', 
    'snowflake', 'optimization', 'git', 
    'powerpoint', 'nosql', 'data quality', 
    'data visualization', 'data mining', 
    'aws', 'business intelligence', 
    'data engineer', 'agile', 'java', 
    'rest', 'query', 'data analyst', 
    'python', 'data scientist', 
    'project management', 'big data', 
    'data modeling', 'azure', 'api', 
    'unity'
    ]

skills_vec = dataset_skills["skills_vec"].apply(literal_eval).apply(pd.Series)[skills_lst]
```

```python
len(skills_lst)
```

```python
skills_vec_binary = skills_vec.copy()
threshold = 0.9
skills_vec_binary[skills_vec_binary > threshold] = 1
skills_vec_binary[skills_vec_binary < threshold] = 0
```

```python
skills_vec_binary
```

```python
knn = NearestNeighbors(n_neighbors=5, algorithm='brute', metric='cosine', n_jobs=-1)
knn.fit(skills_vec_binary)
print("ok")
```

```python
def recommend_jobs(skills_lst, skills = {}):
  initiate_arr = [0] * len(skills_lst)
  idx = []
  for s in skills:
    i = skills_lst.index(s)
    idx.append(i)
  for i in idx:
    initiate_arr[i] = 1
  distances, indices = knn.kneighbors([initiate_arr])
  return dataset_skills.iloc[indices[0]]
```

```python
recommend_jobs(skills_lst, skills = {'sql', 'tableau','data visualization','hadoop', 'python'})
```
